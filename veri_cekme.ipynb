```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = "https://www.emlakjet.com/kiralik-daire/duzce"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html")
```

```python
explanations = soup.find_all("h3", class_="styles_title__CN_n3")
l_explanation = []

for expalanation in explanations:
    l_explanation.append(expalanation.get_text())
l_explanation = np.array(l_explanation)
l_explanation
```

```python
locations = soup.find_all("span", class_="styles_location__ieVpH")
l_locations = []
for location in locations:
    l_locations.append(location.get_text())

l_locations
```

```python
def data_scraping(tag:str, attr:str ,attr_value:str) -> np.array:
    data_list = []
    if attr == "class":
        data = soup.find_all(f"{tag}", class_ = f"{attr_value}")
    for i in data:
        data_list.append(i.get_text())

    return data_list
```

```python
print(data_scraping("span", "class","styles_location__ieVpH"))
```
