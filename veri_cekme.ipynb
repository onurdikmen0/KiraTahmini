```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
page = 2
all_data = []

def data_scraping(tag:str, attr:str ,attr_value:str) -> np.array:
    data_list = []
    if attr == "class_":
        data = soup.find_all(f"{tag}", class_ = f"{attr_value}")
    for i in data:
        data_list.append(i.get_text())
    return data_list


while page < 8:
    url = f"https://www.emlakjet.com/kiralik-daire/duzce?sayfa={page}"
    print(f"Sayfa {page} -> {url}")

    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html")

   
    
    st2 = soup.find_all("div", attrs= {"styles_listingWrapper__I0H_l"})
    if not st2: # veri gelmediyse duracak
        break


    locations = data_scraping("span", "class_","styles_location__ieVpH")
    # explanations = data_scraping("h3", "class_","styles_title__CN_n3")
    informations = data_scraping("div", "class_" ,"styles_quickinfoWrapper__F5BBD")
    sales_prices = data_scraping("div", "class_", "styles_priceContent__TBOLV")
    
    # her sayfayı ekler
    for loc, info, price in zip(locations, informations, sales_prices):
        all_data.append({
        "location": loc,
        "information": info,
        "sales_price": price
    })


    page += 1

url = f"https://www.emlakjet.com/kiralik-daire/duzce"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html")
st2 = soup.find_all("div", attrs= {"styles_listingWrapper__I0H_l"})
df1 = pd.DataFrame(all_data)
df1.head()
```
