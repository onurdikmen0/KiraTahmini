Tüm CSV leri tek bir CSV yap.

```python
import pandas as pd
import glob
import os
```

```python

# CSV dosyalarının bulunduğu klasör
folder_path = "kira"

# Klasördeki tüm CSV dosyalarını bul
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Tüm CSV'leri okuyup birleştir
df_list = []
for file in csv_files:
    df = pd.read_csv(file)
    df_list.append(df)

# DataFrame'leri birleştir
merged_df = pd.concat(df_list, ignore_index=True)

# Tek CSV olarak kaydet
merged_df.to_csv("turkey_rent.csv", index=False)

print(f"{len(csv_files)} dosya birleştirildi")
```
