```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

Şehirlerin kaç sayfa ilanı olduğunu buluyor sonuçları ise dictionary de saklıyor. 

```python
cities = [
    "adana", "adiyaman", "afyonkarahisar", "agri", "aksaray", "amasya", "ankara", "antalya",
    "ardahan", "artvin", "aydin", "balikesir", "bartin", "batman", "bayburt", "bilecik",
    "bingol", "bitlis", "bolu", "burdur", "bursa", "canakkale", "cankiri", "corum",
    "denizli", "diyarbakir", "duzce", "edirne", "elazig", "erzincan", "erzurum",
    "eskisehir", "gaziantep", "giresun", "gumushane", "hakkari", "hatay", "igdir",
    "isparta", "mersin", "istanbul", "izmir", "kahramanmaras", "karabuk", "karaman",
    "kars", "kastamonu", "kayseri", "kilis", "kirikkale", "kirklareli", "kirsehir",
    "kocaeli", "konya", "kutahya", "malatya", "manisa", "mardin", "mugla", "mus",
    "nevsehir", "nigde", "ordu", "osmaniye", "rize", "sakarya", "samsun", "sanliurfa",
    "siirt", "sinop", "sivas", "tekirdag", "tokat", "trabzon", "tunceli", "usak",
    "van", "yalova", "yozgat", "zonguldak", "sirnak"
]
cities_pages = {}

for city in cities:
    url = f"https://www.emlakjet.com/kiralik-konut/{city}"

    print(f"Sayfa {city} -> {url}")

    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

   
    page_numbers = []
    for li in soup.select("ul.styles_list__ExTZw li"):
        text = li.get_text(strip = True)
        if text.isdigit():
            page_numbers.append(int(text))

    if page_numbers:
        total_pages = max(page_numbers)
        print("toplam sayfa sayısı",total_pages)
    else:
        total_pages = 1
        print("Toplam sayfa sayısı",total_pages)

    cities_pages[city] = total_pages
```

Şehirleri ve sayfa sayılarını listeler

```python
print("\nTüm şehirlerin sayfa sayıları:")
print(cities_pages)
```

```python
# Şehir listesi
cities = [
    "adana", "adiyaman", "afyonkarahisar", "agri", "aksaray", "amasya", "ankara", "antalya",
    "ardahan", "artvin", "aydin", "balikesir", "bartin", "batman", "bayburt", "bilecik",
    "bingol", "bitlis", "bolu", "burdur", "bursa", "canakkale", "cankiri", "corum",
    "denizli", "diyarbakir", "duzce", "edirne", "elazig", "erzincan", "erzurum",
    "eskisehir", "gaziantep", "giresun", "gumushane", "hakkari", "hatay", "igdir",
    "isparta", "mersin", "istanbul", "izmir", "kahramanmaras", "karabuk", "karaman",
    "kars", "kastamonu", "kayseri", "kilis", "kirikkale", "kirklareli", "kirsehir",
    "kocaeli", "konya", "kutahya", "malatya", "manisa", "mardin", "mugla", "mus",
    "nevsehir", "nigde", "ordu", "osmaniye", "rize", "sakarya", "samsun", "sanliurfa",
    "siirt", "sinop", "sivas", "tekirdag", "tokat", "trabzon", "tunceli", "usak",
    "van", "yalova", "yozgat", "zonguldak", "sirnak"
]

all_data = []


def data_scraping(soup, tag: str, attr_value: str):
    """
    Fonksiyon: soup üzerinden veri çekme
    """
    data_list = [i.get_text().strip() for i in soup.find_all(tag, class_=attr_value)]
    return data_list


def scrape_page(city, page_num=None):
    """
    # Fonksiyon: tek bir sayfadan ilanları çek
    """
    if page_num is None:
        url = f"https://www.emlakjet.com/kiralik-daire/{city}"
    else:
        url = f"https://www.emlakjet.com/kiralik-daire/{city}?sayfa={page_num}"
    
    try:
        response = requests.get(url)
        if response.status_code != 200:
            print(f"Failed to fetch {url} | Status code: {response.status_code}")
            return []

        soup = BeautifulSoup(response.content, "html.parser")
        
        locations = data_scraping(soup, "span", "styles_location__ieVpH")
        informations = data_scraping(soup, "div", "styles_quickinfoWrapper__F5BBD")
        sales_prices = data_scraping(soup, "div", "styles_priceContent__TBOLV")

        # Verilerin eşleşip eşleşmediğini kontrol et
        if not (len(locations) == len(informations) == len(sales_prices)):
            print(f"Data length mismatch on {city} page {page_num}")
        
        page_data = [
            {"location": loc, "information": info, "sales_price": price}
            for loc, info, price in zip(locations, informations, sales_prices)
        ]
        return page_data
    
    except Exception as e:
        print(f"Error processing {url}: {e}")
        return []

# Tüm şehirler ve sayfalar üzerinde döngü
for city, total_pages in cities_pages.items():
    if total_pages > 1:
        for page_num in range(1, total_pages + 1):
            all_data.extend(scrape_page(city, page_num))
    else:
        all_data.extend(scrape_page(city))

# DataFrame oluştur
df1 = pd.DataFrame(all_data)
print(df1.head())
```
