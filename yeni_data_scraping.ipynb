```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

Şehirlerin kaç sayfa ilanı olduğunu buluyor sonuçları ise dictionary de saklıyor. 

```python
cities = [
    "adana", "adiyaman", "afyonkarahisar", "agri", "aksaray", "amasya", "ankara", "antalya",
    "ardahan", "artvin", "aydin", "balikesir", "bartin", "batman", "bayburt", "bilecik",
    "bingol", "bitlis", "bolu", "burdur", "bursa", "canakkale", "cankiri", "corum",
    "denizli", "diyarbakir", "duzce", "edirne", "elazig", "erzincan", "erzurum",
    "eskisehir", "gaziantep", "giresun", "gumushane", "hakkari", "hatay", "igdir",
    "isparta", "mersin", "istanbul", "izmir", "kahramanmaras", "karabuk", "karaman",
    "kars", "kastamonu", "kayseri", "kilis", "kirikkale", "kirklareli", "kirsehir",
    "kocaeli", "konya", "kutahya", "malatya", "manisa", "mardin", "mugla", "mus",
    "nevsehir", "nigde", "ordu", "osmaniye", "rize", "sakarya", "samsun", "sanliurfa",
    "siirt", "sinop", "sivas", "tekirdag", "tokat", "trabzon", "tunceli", "usak",
    "van", "yalova", "yozgat", "zonguldak", "sirnak"
]
cities_pages = {}

for city in cities:
    url = f"https://www.emlakjet.com/kiralik-konut/{city}"

    print(f"Sayfa {city} -> {url}")

    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

   
    page_numbers = []
    for li in soup.select("ul.styles_list__ExTZw li"):
        text = li.get_text(strip = True)
        if text.isdigit():
            page_numbers.append(int(text))

    if page_numbers:
        total_pages = max(page_numbers)
        print("toplam sayfa sayısı",total_pages)
    else:
        total_pages = 1
        print("Toplam sayfa sayısı",total_pages)

    cities_pages[city] = total_pages
```

Şehirleri ve sayfa sayılarını listeler

```python
print("\nTüm şehirlerin sayfa sayıları:")
print(cities_pages)
```

```python

sayfa_sayisi = 7
sehir = "yalova"


# tag ve attribute parametrelerini verince ilgili verileri çekiyor.
def data_scraping(tag:str, attr_value:str) -> np.array:
    data_list = []
    data = soup.find_all(f"{tag}", class_ = f"{attr_value}")
    for i in data:
        data_list.append(i.get_text())
    return data_list

all_data = []

for i in range(2, sayfa_sayisi + 1):
    url = f"https://www.emlakjet.com/kiralik-daire/{sehir}?sayfa={i}"


    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    a = soup.find_all("div", attrs= {"styles_listingWrapper__I0H_l"})

    locations = data_scraping("span", "styles_location__ieVpH")
    informations = data_scraping("div", "styles_quickinfoWrapper__F5BBD")
    sales_prices = data_scraping("div", "styles_priceContent__TBOLV")
    
    # her sayfayı ekler
    for loc, info, price in zip(locations, informations, sales_prices):
        all_data.append({
        "location": loc,
        "information": info,
        "sales_price": price
    })

# 1. sayfa hariç sayfalar için
df1 = pd.DataFrame(all_data)
```

CSV olarak indir

```python
sehir = "yalova"
df1.to_csv(f"{sehir}_data.csv", index=False)
```
