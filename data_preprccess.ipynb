```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("turkiye_kiralik_daireler.csv")
df.head()
```

information kolonunu parçalıyorum

```python
df[["konut_tipi","oda_sayisi","kat","metrekare"]] = df["information"].str.split("|", expand = True).apply(lambda x : x.str.strip())
```

```python
df.head()
```

birden fazla veri olan information kolonunu siliyorum. Artık "konut_tipi","oda_sayisi","kat","metrekare" var

```python
df.drop("information", axis=1, inplace=True)
```

kat kolonundaki . ları sildim

```python
df["kat"] = df["kat"].str.replace(".", "", regex=False)
```

kat kolonundaki Kat kelimesini sildim

```python
df["kat"] = df["kat"].str.replace("Kat", "", regex=False).str.strip()
```

```python
df.head()
```

```python
df["metrekare"] = df["metrekare"].str.replace("m²", "", regex=False).str.strip()
df.head()
```

```python
df["sales_price"] = df["sales_price"].str.replace("TL", "", regex=False)
df["sales_price"] = df["sales_price"].str.replace(".", "", regex=False)
df["sales_price"] = df["sales_price"].str.strip()

df.head()
```

```python
df.rename(columns={"sales_price" : "fiyat"}, inplace=True)
df.head()
```

```python
df[["il","ilce"]] = df["location"].str.split(" - ",expand=True)
df.drop(columns=['location'], inplace=True)
df.head()
```

```python
# df["mahalle"] = df["mahalle"].astype(str)
# df["mahalle"] = df["mahalle"].str.replace("Mahallesi", "").str.strip()
```

```python
new_sort = ["konut_tipi","oda_sayisi","kat","metrekare","il","ilce","fiyat"]
df = df[new_sort]
df
```

```python
df["il"].dtype
```

```python
df["il"] = df["il"].astype("string")
df["ilce"] = df["ilce"].astype("string")
df["konut_tipi"] = df["konut_tipi"].astype("string")
```
